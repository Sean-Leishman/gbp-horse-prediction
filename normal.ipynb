```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
rd = pd.read_csv("data/raw/processed_full_data_with_features.csv", index_col=[0])
fd = pd.read_csv("data/raw/processed_full_data_with_horse_history.csv", index_col=[0])
```

```python
max_days_since_last_race = max(rd['days_since_last_race'])
rd.loc[rd.days_since_last_race == 0, 'days_since_last_race'] = max_days_since_last_race
```

```python
max_days_since_last_race = max(fd['days_since_last_race'])
fd.loc[rd.days_since_last_race == 0, 'days_since_last_race'] = max_days_since_last_race
```

```python
rd = rd.drop(['placed', 'places', 'won', 'going__firm_win_percent', 'going__good_win_percent', 'going__heavy_win_percent'], axis=1)
fd = fd.drop(['placed','places','won'], axis=1)
```

```python
fd_scale_df_columns = fd[['horse_weight', 'horse_win_percents', 'jockey_win_percent', 'trainer_win_percent', 'days_since_last_race', 'mean_speed_figures', 'last_figures', 'best_figures_dist', 'best_figures_going', 'top_speeds', 'ratings', 'official_ratings', 'odds', 'days', 'mean_ratings', 'last_ratings', 'last_official_rating', 'difference_in_ratings', 'best_rating_going', 'best_official_rating_going', 'win_percent_going', 'best_rating_dist', 'best_official_rating_dist', 'win_percent_dist', 'length', 'sire_win_percent', 'dam_win_percent', 'dam_sire_win_percent', 'sire_og_going_win_percent', 'sire_prog_going_win_percent', 'sire_og_type_win_percent', 'sire_prog_type_win_percent', 'sire_og_dist_win_percent', 'sire_prog_dist_win_percent', 'dam_og_going_win_percent', 'dam_prog_going_win_percent', 'dam_og_type_win_percent', 'dam_prog_type_win_percent', 'dam_og_dist_win_percent', 'dam_prog_dist_win_percent']]
rd_scale_df_columns = rd[['horse_weight', 'horse_win_percents', 'jockey_win_percent', 'trainer_win_percent', 'days_since_last_race', 'mean_speed_figures', 'last_figures', 'best_figures_dist', 'best_figures_going', 'top_speeds', 'ratings', 'official_ratings', 'odds', 'days', 'mean_ratings', 'last_ratings', 'last_official_rating', 'difference_in_ratings', 'best_rating_going', 'best_official_rating_going', 'win_percent_going', 'best_rating_dist', 'best_official_rating_dist', 'win_percent_dist', 'length', 'sire_win_percent', 'dam_win_percent', 'dam_sire_win_percent', 'sire_og_going_win_percent', 'sire_prog_going_win_percent', 'sire_og_type_win_percent', 'sire_prog_type_win_percent', 'sire_og_dist_win_percent', 'sire_prog_dist_win_percent', 'dam_og_going_win_percent', 'dam_prog_going_win_percent', 'dam_og_type_win_percent', 'dam_prog_type_win_percent', 'dam_og_dist_win_percent', 'dam_prog_dist_win_percent']]
```

```python
scaler = StandardScaler()
```

```python
rd_scaled = pd.DataFrame(data=scaler.fit_transform(rd_scale_df_columns), index=rd_scale_df_columns.index, columns=rd_scale_df_columns.columns)
fd_scaled = pd.DataFrame(data=scaler.fit_transform(fd_scale_df_columns), index = fd_scale_df_columns.index, columns=fd_scale_df_columns.columns)
```

```python
rd.update(rd_scaled)
```

```python
fd.update(fd_scaled)
```

```python
from sklearn.model_selection import train_test_split

fd_train_data, fd_test_data = train_test_split(fd_scaled)
```
