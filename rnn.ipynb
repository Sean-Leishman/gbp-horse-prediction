```python
class HorseHistoryDataset(Dataset):
    def __init__(self, raw_data_file , filtered_data_file,  index_file, transform=None) -> None:
        self.raw_data = pd.read_csv(raw_data_file, index_col=[0])
        self.filtered_data = pd.read_csv(filtered_data_file, index_col=[0])
        self.horse_history_index = pd.read_csv(index_file, index_col=[0])

        self.raw_data = self.raw_data.set_index(["date_race_id", "horse_ids"])
        self.filtered_data = self.filtered_data.set_index(["date_race_id", "horse_ids"])
        self.horse_history_index = self.horse_history_index.set_index(["race_id", "horse_ids"])   

        print(self.raw_data.head())
        print(self.filtered_data.head())
        print(self.horse_history_index.head())

        self.transform=transform

    def __len__(self) -> int:
        return len(self.raw_data)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        horse = self.filtered_data.iloc[idx]
        if self.transform:
            pass 

        try: 
            print(idx, horse.race_id, horse.name)
            hist = self.horse_history_index.loc[(horse.race_id, horse.name[1])]
            print(hist)
            print(hist.date_race_id.values, horse.name[1])
            x_data = torch.tensor(self.raw_data.loc[(hist.date_race_id.values[0], horse.name[1])].values.astype('float64'))
            print(x_data)
            # Gets the horse race that is currently occuring. Will be predicting the track stats too
            # Could just extract 'won' variable
            y_data = torch.tensor(horse.values.astype('float64'))
            print(y_data)
        except KeyError as e:
            print("Key Error")
            return None
        
        return x_data, y_data
    
tl = DataLoader(HorseHistoryDataset("data/raw/processed_full_data_with_features.csv", "data/raw/processed_full_data_with_horse_history.csv", "data/raw/fff.csv"), batch_size=32)
```
