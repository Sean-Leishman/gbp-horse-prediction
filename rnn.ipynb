```python
from torch.utils.data import Dataset, DataLoader
import torch
```

```python
class HorseHistoryDataset(Dataset):
    def __init__(self, raw_data_file , index_file, transform=None) -> None:
        self.raw_data = pd.read_csv(raw_data_file, index_col=[0])
        self.horse_history_index = pd.read_csv(index_file, index_col=[0])

        self.raw_data = self.raw_data.set_index(["date_race_id", "horse_ids"])
        self.horse_history_index = self.horse_history_index.set_index(["race_id", "horse_ids"])   

        self.transform=transform

    def __len__(self) -> int:
        return len(self.raw_data)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()
        
        print(idx)

        horse = self.raw_data.iloc[idx]
        if self.transform:
            pass 

        try: 
            hist = self.horse_history_index.loc[(horse.race_id, horse.index[1])]
            x_data = torch.tensor(self.raw_data.loc[(hist.data_race_id, horse.horse_ids)].values)
            # Gets the horse race that is currently occuring. Will be predicting the track stats too
            # Could just extract 'won' variable
            y_data = torch.tensor(horse).values
        except KeyError as e:
            return None
        
        return x_data, y_data
    
tl = DataLoader(HorseHistoryDataset("data/raw/full_data_with_horse_history.csv", "fff.csv"), batch_size=32)
```
